```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(1,5, 21)
y = x**3
y
```

```python
fig = plt.figure(dpi=100)

axes = fig.add_axes([0.0,0.0,1,1])
axes.grid(True)

axes2 = fig.add_axes([0.01,0.5,0.49,0.49])
axes2.grid(True)
axes2.set_facecolor('black')

axes.plot(x,y, 'm:^')
axes2.hist(y)

plt.show()
```

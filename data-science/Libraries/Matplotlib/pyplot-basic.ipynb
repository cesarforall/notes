## Pyplot
Es una herramienta de matplotlib que permite ejecutar gráficas de manera sencilla

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,10,20)
y = x**3
```

```python
x
```

## .plot
Gráfico de líneas

```python
plt.plot(x,y)
plt.show()
```

El tercer parámetro cambia los colores
- b = blue
- y = yellow
- r = r
- m = magenta

```python
plt.plot(x,y,'r')
plt.show()
```

También se pueden cambiar los markers a:
- o = círculos
- v = flechas
- x = equis
- s = squares
- D = diamantes
- '.' = punto
- ',' = pixel
- '^' = triangle up
- '<' = triangle left
- '>' = triangle right

Estilos de línes
- '-' = continuidad en el marker
- '--' = dashed
- '-.' = dash dot
- ':' = línea de puntos

```python
plt.plot(x, y,'bs-')
plt.show()
```

```python
plt.plot(x, y,'gD:')
plt.show()
```

## .hist()
Gráfico histograma para una sola variables

```python
plt.hist(x)
plt.show()
```

## .pie()
Gráfico pie para una sola variables

```python
plt.pie(x)
plt.show()
```

## .scatter()
Gráfico scatter para encontrar la relación entre variables distintas

```python
plt.scatter(x,y)
plt.show()
```

## .boxplot()


```python
plt.boxplot(x)
plt.show()
```

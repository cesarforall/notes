### Subplot
Subplot es una función de Python que permite trabajar con más de una gráfica a la vez. Su funcionamiento consiste en definir un número de filas, columnas y un índice antes de de la gráfica a representar para marcar su posición en la pantalla según el grid.

    plt.subplot(3,3,9)

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
plt.subplot(1,2,1)
plt.plot(x,y)
plt.subplot(1,2,2)
plt.hist(y)
plt.show()
```

```python
z = np.random.randint(1,100,10)
zz = z**2
zz
```

```python
plt.plot(z,zz, ':D')
plt.show()
```

```python
plt.subplot(3,3,1)
plt.bar(z,zz)
plt.subplot(3,3,9)
plt.plot(z,zz)
plt.show()
```

### .head() and .tail()
Las funciones .head(n) and .tail(n) devuelven el número de datos n del principio o el final del DataFrame. Si no se especifica un argumento, n vale 5 por defecto.

```python
df_results_survey_2022.tail(3)
```

### .memory_usage()
La función .memory_usage() devuelve el peso en memoria de los datos. La opción deep=True especifica mejor precisión en la medida de los datos

```python
df_results_survey_2022.memory_usage(deep=True)
```

### .drop_duplicates()
La función .drop_duplicates() elimina las filas duplicadas. La opción keep='last' elimina el primer resutado que encuentre


```python
newData = df_results_survey_2022.loc[0]
```

```python
df_results_survey_2022 = df_results_survey_2022.append(newData)
```

```python
df_results_survey_2022.head(2)
```

```python
df_results_survey_2022.drop_duplicates(keep='last').head(2)
```

### .sort_values()
La función .sort_values() ordena los valores según una columna. La opción ascending=False devuelve los datos de forma descendiente

```python
questions = df_results_survey_2022.columns.values.tolist()
type(questions)

for question in questions:
    print(question)
```

```python
df_results_survey_2022.sort_values('WebframeWantToWorkWith', ascending=False)
```

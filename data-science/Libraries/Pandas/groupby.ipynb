## .groupby()


La funcin .groupby() agrupa los valores repetido de una columna. La columna se puede usar como ndice para acceder por medio de .loc

```python
import pandas as pd
```

```python
df_books = pd.read_csv('./bestsellers-with-categories.csv')
df_books.head(2)
```

### Funciones que pueden acompaar a .groupby()
- .count() cuenta los registros
- .sum() suma los registros
- .max() devuelve el valor mximo de los registros

```python
df_books.groupby('Price').sum()
```

```python
df_books.groupby('User Rating').count()
```

###
La funcin .reset_index() libera la columna como ndice

```python
df_books.groupby('User Rating').count().reset_index()
```

###
La funcin .agg() puede devolver varias mtricas distintas

```python
df_books.groupby('Author').agg(['min', 'max'])
```

```python
df_books.head(2)
```

Agrupar con varios ndices y tipo de evaluacin

```python
df_books.groupby('Author').agg({'Reviews':['min', 'max'], 'User Rating': 'sum'})
```

Agrupar por varios ndices

```python
df_books.groupby(['Author', 'Year']).count()
```

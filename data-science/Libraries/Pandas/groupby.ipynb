## .groupby()


La función .groupby() agrupa los valores repetido de una columna. La columna se puede usar como índice para acceder por medio de .loc

```python
import pandas as pd
```

```python
df_books = pd.read_csv('./bestsellers-with-categories.csv')
df_books.head(2)
```

### Funciones que pueden acompañar a .groupby()
- .count() cuenta los registros
- .sum() suma los registros
- .max() devuelve el valor máximo de los registros

```python
df_books.groupby('Price').sum()
```

```python
df_books.groupby('User Rating').count()
```

###
La función .reset_index() libera la columna como índice

```python
df_books.groupby('User Rating').count().reset_index()
```

###
La función .agg() puede devolver varias métricas distintas

```python
df_books.groupby('Author').agg(['min', 'max'])
```

```python
df_books.head(2)
```

Agrupar con varios índices y tipo de evaluación

```python
df_books.groupby('Author').agg({'Reviews':['min', 'max'], 'User Rating': 'sum'})
```

Agrupar por varios índices

```python
df_books.groupby(['Author', 'Year']).count()
```

### .apply()
Aplica una función a los datos de la columna de un dataframe. Es más rápido que aplicar un ciclo for

```python
df_books = pd.read_csv('./bestsellers-with-categories.csv')
```

```python
df_books.head(2)
```

```python
def two_times(value):
    return value * 2
```

```python
df_books['Price'].apply(two_times)
```

```python
df_books['New Price'] = df_books['Price'].apply(two_times)
```

```python
df_books.head()
```

```python
df_books['Third Price'] = df_books['Price'].apply(lambda x: x * 3)
```

```python
df_books.tail()
```

Para tratar los datos en todo el dataframe usando apply se especifica el axis

```python
df_books.columns.values
```

```python
df_books['New Rating'] = df_books.apply(lambda x: x['User Rating'] + 100 if x['Year'] == 2019 else x['Price'], axis=1)
```

```python
import pandas as pd
```

```python
df_books = pd.read_csv('./bestsellers-with-categories.csv')
df_books.columns
```

### .pivot_table()
- index='column' define una columna como índice
- columns='column' define una columna como nueva columa
- value='colum' define el valor de las nuevas columnas
- aggfunc='function' define una función para tratar los valores

```python
df_books.pivot_table(index='Author', columns='Genre', values='Price', aggfunc='mean')
```

### .melt()
Devuelve filas por cada dato en una columna

```python
df_books[['Author', 'Year']].melt()
```

### .melt()
- id_vars='column' especifica la columna que no se hará melt
- value_vars='column' especifica la columna a representar con meelt

```python
df_books.melt(id_vars='Author', value_vars='Genre')
```

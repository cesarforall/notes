## .drop()
La función .drop() de pandas elimina las columnas o filas de un data frame.
- El primer parámetro indica el índice
- La opción axis=# especifica si es fila o columna
- La opción inplace=True define si se aplica al archivo original

    dt_pandas = dt_pandas.drop('Columna', axis=1)

- Para aplicar el cambio en el archivo original sin la opción inplace se reasigna la variable
    
    dt_pandas = dt_pandas.drop('Columna', axis=1)

- También se puede usar la función 'del' de python pero no es recomendable
    
    del dt_pandas['Columna']
        

### Eliminar columnas de un dataframe con pandas

```python
df_books_no_reviews = df_books.drop('Reviews', axis=1)
```
